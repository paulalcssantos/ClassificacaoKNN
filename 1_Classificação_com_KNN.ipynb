#KNN

*O algoritmo KNN (K-Nearest Neighbors) é um algoritmo de aprendizado de máquina supervisionado usado para classificação e regressão. Ele funciona classificando um novo ponto de dados com base na classe dos seus k vizinhos mais próximos no conjunto de dados de treinamento.*

Será utilizado o famoso Iris Dataset na aplicação do modelo:

https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris

##Separando os dados

**1. load_iris(as_frame=True):**

    A função load_iris carrega o conjunto de dados Iris.
    O parâmetro as_frame=True indica que os dados devem ser retornados como um pandas DataFrame, o que facilita a manipulação e análise dos dados.

**2. ['data']:**

    O conjunto de dados Iris retornado por load_iris é um objeto do tipo Bunch, que é semelhante a um dicionário.
    A chave 'data' acessa o DataFrame que contém as características (features) das amostras de flores. Estas características são: comprimento da sépala, largura da sépala, comprimento da pétala e largura da pétala.

**3. X = load_iris(as_frame=True)['data']:**

    Aqui, X é atribuído ao DataFrame contendo as características das amostras de flores.
    X terá 150 linhas e 4 colunas, correspondendo às 150 amostras e suas 4 características.

**4.['target']:**

    A chave 'target' acessa a série (ou coluna) que contém os rótulos das classes para cada amostra. Estes rótulos representam os tipos de íris (Iris-setosa, Iris-versicolor, e Iris-virginica), codificados como 0, 1 e 2, respectivamente.

**5. y = load_iris(as_frame=True)['target']:**

    Aqui, y é atribuído à série contendo os rótulos das classes das amostras de flores.
    y terá 150 valores, um para cada amostra.

```python
#Características da pétala
X = load_iris(as_frame=True)['data']

#Tipo de íris
y = load_iris(as_frame=True)['target']
```

###Treinamento e Teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
#Algoritmo sem dados nenhum, onde K = 3

knn = KNeighborsClassifier(n_neighbors=3)
```

```python
#Treinando o algoritmo

knn.fit(X_train, y_train)
```

### Fazendo previsões

```python
#Usando o algoritmo já treinado para fazer previsões usando os dados de teste

y_pred = knn.predict(X_test)

y_pred
```

```python
# Verificando a quantidade de casos errados

print("Número de íris que previmos errado de um total de {0} observações de teste: {1}"\
      .format(X_test.shape[0], (y_test != y_pred).sum()))
```

##Deixando os dados com a mesma média e mesma variância

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import pandas as pd
```

```python
pd.DataFrame(X_train).describe()
```

```python
#algoritmo sem nenhum dado, onde K = 3

knn = KNeighborsClassifier(n_neighbors=3)
```

```python
#Treinando o algoritmo

knn.fit(X_train, y_train)
```

```python
#Realizando as previsões

y_pred = knn.predict(X_test)

y_pred
```

```python
# Verificando a quantidade de casos errados

print("Número de íris que previmos errado de um total de {0} observações de teste: {1}"\
      .format(X_test.shape[0], (y_test != y_pred).sum()))
```

##Calculando o melhor valor de K

```python
percentual_acerto = []
tamanho_teste = X_test.shape[0]

for i in range(1, 15):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, y_train)
  pred_i = knn.predict(X_test)

  acerto_i = (y_test == pred_i).sum()/tamanho_teste
  percentual_acerto.append(100 * acerto_i)
```

```python
import pandas as pd
```

```python
df_acerto = pd.DataFrame()
df_acerto['K'] = range(1, 15)
df_acerto['percetual_acerto'] = percentual_acerto

df_acerto.head()
```

##Gerando gráfico

```python
import plotly.express as px
```

```python
fig = px.line(df_acerto, x='K', y='percetual_acerto',
              title="Percentual de Acerto por K")
fig.show()
```
